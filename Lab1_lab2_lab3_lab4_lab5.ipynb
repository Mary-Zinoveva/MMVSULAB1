Построение математической модели у'=kу

```
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

#Определяем функцию, задающую значения параметрам дифференциального уравнения, определяем вид диффура
def dydt(y,t):
  k=0.05
  return k*y

#Создаем область построения
t=np.linspace(0,100,100)

#решаем диффур, обозначаем функцию решения solve
solve=odeint(dydt, y0=1, t=t)
solve=solve.flatten()

#Строим график мат.модели
%matplotlib inline
plt.plot(t, solve, label=r"$у'=kу$")
plt.grid(True)
plt.legend()
plt.show()
```

# Лабораторная работа №4

Построение математической модели с использованием своей функции поиска решения, основанной на методе Эйлера

```
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
n=1000
h=0.01
#Определяем функцию, задающую значения параметрам дифференциального уравнения, определяем вид диффура
def dydt(y,t):
  k=0.42
  return k*y

def eiler(y0,t0,h,n):
  f=np.zeros((2,n+1))
  f[0][0]=y0
  f[1][0]=t0
  for i in range(n):
    f[0][i+1]=f[0][i]+h*dydt(f[0][i],f[1][i])
    f[1][i+1]=f[1][i]+h
    i=i+1
  return f

#решаем диффур, обозначаем функцию решения solve
solve1=eiler(y0=0.1,t0=0,h=h,n=n)
t=np.arange(0,10+h,h)
t=np.linspace(0,10,n)
solve2=odeint(dydt, y0=0.1, t=t)
#solve=solve.flatten()

#Строим график мат.модели
%matplotlib inline
plt.plot(solve1[1], solve1[0], label=r"y'=ky",color="black")
plt.plot(t, solve2,color="red")
plt.grid(True)
plt.legend()
plt.show()
```

# Лабораторная работа №5

Построение некоторых математических моделей с помощью встроенных функций

```
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,1,100)
y=x+x*np.random.random(len(x))
A=np.vstack([x,np.ones(len(x))]).T
a,b=np.linalg.lstsq(A,y,rcond=None)[0]

plt.plot(x,y,'.')
plt.plot(x,x*a+b,'r')
plt.grid(True)
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

x=np.linspace(-1,1,100)
y=x**2+x*np.random.random(len(x))
def f(x,a,b,c):
   return a*x**2+b*x+c

(a,b,c),_=curve_fit(f,x,y)

plt.plot(x,y,'.')
plt.plot(x,f(x,a,b,c),'r')
plt.grid(True)
plt.show()
```
