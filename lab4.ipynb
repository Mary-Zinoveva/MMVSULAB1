```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

#Определяем функцию, задающую значения параметрам дифференциального уравнения, определяем вид диффура
def dydt(y,t):
  k=0.0012
  return -k*y+t
p=100
n=20
h=p/n
t0=0


def eiler(y0,t0,h):
  f=np.zeros((2,n+1))
  f[0][0]=y0
  f[1][0]=t0
  for i in range(n):
    f[0][i+1]=f[0][i]+h*dydt(f[0][i],f[1][i])
    f[1][i+1]=f[1][i]+h
    i=i+1
  return f

#решаем диффур, обозначаем функцию решения solve
solve1=eiler(y0=3000,t0=t0,h=h)

t=np.linspace(0,p,n)
solve2=odeint(dydt, y0=3000, t=t)
#solve=solve.flatten()

#Строим график мат.модели
%matplotlib inline
plt.plot(solve1[1], solve1[0], label=r"$x'=kx$")
plt.plot(t, solve2,color="green")
plt.grid(True)
plt.legend()
plt.show()
```
